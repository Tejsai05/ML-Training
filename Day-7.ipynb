```python
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
average_profit_margin_by_product = df.groupby('Product')['Profit Margin'].mean().reset_index()
average_profit_margin_by_product.columns = ['Product', 'Average Profit Margin']
print(average_profit_margin_by_product)
```
