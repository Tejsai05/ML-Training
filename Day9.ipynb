```python
account_type_summary = data.groupby('Account_Type').agg(
    Total_Transaction_Amount=('Transaction_Amount', 'sum'),
    Average_Account_Balance=('Account_Balance', 'mean')
).reset_index()

print("Summary by Account Type:")
print(account_type_summary)
```

```python
branch_summary = data.groupby('Branch').agg(
    Total_Transactions=('Transaction_Amount', 'count'),
    Average_Transaction_Amount=('Transaction_Amount', 'mean')
).reset_index()

print("\nSummary by Branch:")
print(branch_summary)
```
