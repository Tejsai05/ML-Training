```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("day_9_banking_data.csv")
data
```

```python
total_transaction_per_account_type = data.groupby('Account_Type')['Transaction_Amount'].sum()
total_transaction_per_account_type.plot(kind='bar', title='Total Transaction Amount per Account Type')

```

```python
transaction_per_branch = data['Branch'].value_counts()
transaction_per_branch.plot(kind='pie', autopct='%1.1f%%', title='Percentage of Transactions per Branch')
plt.show()
```
