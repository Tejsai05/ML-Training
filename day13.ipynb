```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("day_13_Pharma_data.csv")
data
```

```python
data.isnull()
```

```python
plt.figure(figsize=(10, 6))
sales_per_region = data.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region', y='Sales', data=sales_per_region)
plt.title('Total Sales per Region')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=data)
plt.title('Relationship between Marketing Spend and Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_Group', y='Effectiveness', data=data)
plt.title('Drug Effectiveness across Age Groups')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sales_trend = data.groupby(['Trial_Period', 'Product_Name'])['Sales'].sum().reset_index()
sns.lineplot(x='Trial_Period', y='Sales', hue='Product_Name', data=sales_trend)
plt.title('Sales Trend for Each Product over Trial Periods')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = data[['Sales', 'Marketing_Spend', 'Effectiveness']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
