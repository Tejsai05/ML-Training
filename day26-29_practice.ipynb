```python
from sklearn.preprocessing import normalize
normalized=normalize(df)
normalized=pd.DataFrame(normalized,columns=df.columns)
normalized
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
dend = sch.dendrogram(sch.linkage(normalized, method='ward'))
plt.title('Customer Dendrogram')
plt.xlabel('Customer Data')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
len(set(dend['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group=AgglomerativeClustering(n_clusters=4)
cluster=group.fit_predict(normalized)
cluster
```

```python
group_num=pd.DataFrame(cluster)
group_num
```

```python
pd.concat([pd.DataFrame(df),group_num],axis=1)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(df,cluster)
```

```python
from sklearn.cluster import  KMeans
wcss=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)#wcss-within cluster sum of states
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

```python
#day29
model=KMeans(n_clusters=5,random_state=309)
groups=model.fit_predict(df)
groups
```
