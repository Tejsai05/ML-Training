```python
from sklearn.preprocessing import normalize
normalized=normalize(df)
normalized=pd.DataFrame(normalized,columns=df.columns)
normalized
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
dend = sch.dendrogram(sch.linkage(normalized, method='ward'))
plt.title('Customer Dendrogram')
plt.xlabel('Customer Data')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
len(set(dend['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group=AgglomerativeClustering(n_clusters=4)
cluster=group.fit_predict(normalized)
cluster
```

```python
group_num=pd.DataFrame(cluster)
group_num
```

```python
pd.concat([pd.DataFrame(normalized),group_num],axis=1)
```
